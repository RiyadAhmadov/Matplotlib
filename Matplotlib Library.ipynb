# 🔎 Matplotlib Library 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

# 💡 Plot

```python
x_array = np.arange(1,6)
y_array = np.arange(3,18,3)
```

```python
plt.plot(x_array,y_array)
```

```python
plt.plot(x_array,y_array,'*')
```

```python
#Let's assign color , marker and size
plt.plot(x_array,y_array,'*:r')
```

```python
#Let's write marker and marker size 
plt.plot(x_array,y_array,marker = '*' , ms = 20)
```

```python
#Let's write marker color and marker border color 
plt.plot(x_array,y_array,marker = 's' , ms = 10,mfc = 'r', mec = 'r')
```

# 📍 Subplot

```python
#Let's draw 2 and more chart in one output

plt.subplot(1,2,1)
plt.plot(y_array)

plt.subplot(1,2,2)
plt.plot(x_array , color = 'r')
```

```python
#Let's write subplot with other way

fig , (plt1 , plt2) = plt.subplots(1,2)
plt1.plot(x_array)
plt2.plot(y_array , color = 'r')
plt1.set_title('Chart 1')
plt2.set_title('Chart 2')
```

# 🔑Line Style 

```python
"""
'solid' 	'-'	
'dotted'	':'	
'dashed'	'--'	
'dashdot'	'-.'	
'None'
"""
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2, 2)

ax1.plot(x_array, y_array, linestyle="dashed", marker="o")
ax1.set_title('Line 1')

ax2.plot(x_array, y_array, linestyle="dotted", marker="o")
ax2.set_title('Line 2')

ax3.plot(x_array, y_array, linestyle="solid", marker="o")
ax3.set_title('Line 3')

ax4.plot(x_array, y_array, linestyle="dashdot", marker="o")
ax4.set_title('Line 4')

plt.tight_layout()

plt.show()
```

```python
#Let's assign width of line
plt.plot(y_array,"*:r",ms = 20 ,linewidth = 5)
```

```python
#Let's draw 2 line in one chart

plt.title("Two line in one chart")
plt.plot(x_array)
plt.plot(y_array)
```

# 🔑 X , Y label 

```python
#Let's write label of x and y arrow 
plt.plot(x_array , y_array , "o--b", mec= "r" ,mfc = 'r')
plt.xlabel("X arrow")
plt.ylabel("Y arrow")
```

# 🔑 Font Properties

```python
#Let's change font of labels
plt.plot(x_array , y_array , "x-.r", mec= "b" ,mfc = 'r' , ms = 15)

font1 = {'family':'cursive', 'color':'darkred', 'size':10}
font2 = {'family':'cursive', 'color':'darkBlue', 'size':20}

plt.title('Arrows' , fontdict = font2)
plt.xlabel("X arrow", fontdict = font1)
plt.ylabel("Y arrow", fontdict = font1)
```

```python
#Let's change location of title 

plt.plot(x_array , y_array , "x-.r", mec= "b" ,mfc = 'r' , ms = 15)
font1 = {'family':'cursive', 'color':'darkBlue', 'size':20}
font2 = {'family':'cursive', 'color':'darkred', 'size':20}
font3 = {'family':'cursive', 'color':'darkgreen', 'size':20}
plt.title('Arrows_l' , fontdict = font1 ,loc = 'left')
plt.title('Arrows_c' , fontdict = font2 ,loc = 'right')
plt.title('Arrows_r' , fontdict = font3 ,loc = 'center')
```

# 📍 Grid

```python
#Let's draw grid in chart
plt.plot(x_array , y_array)
plt.grid()
```

```python
#Let's draw axis x and axis y

fig , (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x_array)
ax1.grid(axis = 'x')
ax2.plot(y_array)
ax2.grid(axis = 'y')
```

```python
#Let's examine properties of grid

plt.plot(x_array , marker = 'o' , linestyle = 'dotted',lw = 5 , ms = 10)
plt.grid(color = 'r' , linestyle = '-.' , linewidth = 1)
```

# 📌 Scatter Plot

```python
#Let's draw scatter plot

z_array = np.arange(10,5,-1)
size = np.arange(30,80,10)

plt.scatter(x_array,y_array , color = 'r' , s = size)
plt.scatter(x_array,z_array)
```

```python
#Let's code for transparency(alpha)

size = [200,23 , 34, 89 , 190]
size1 = [22,100,190,56,240]
plt.scatter(x_array,y_array , color = 'r' , s = size , alpha = 0.9)
plt.scatter(x_array,z_array , color = 'b' , s = size1, alpha = 0.4)
```

# 📌 Bar Plot

```python
#Let's draw bar plot

plt.title("Bar Plot")
plt.bar(x_array ,z_array , color = 'r' , width = 0.2)
```

```python
plt.title('BarHor Plot')
plt.barh(x_array ,z_array , color = 'b' , height = 0.2)
```

# 📌 Histogram

```python
#Let's draw histogram 
values = np.random.normal(200 , 10 , 40)
plt.hist(values , color = 'r')
```

# 📌 Pie Chart

```python
#Let's draw pie chart

names = ['A','B','C','D','E']
plt.pie(x_array , labels = names)
```

```python
#Let's start from 90 angle

plt.pie(x_array , labels = names , startangle = 90)
```

```python
#Let's explode the sides

explode = [0.5,0,0,0,0]
plt.pie(x_array , labels = names , explode = explode)
```

```python
#Let's assign shadow to chart

plt.pie(x_array , labels = names , explode = explode , shadow = True)
```

```python
plt.pie(x_array , labels = names)
plt.legend(title = 'Letters')
plt.show()
```
